```python
data=pd.read_csv("final_past.csv")
```

```python
df_num=data.select_dtypes(exclude="object")## we divide data in this groups to check variance, 
df_object=data.select_dtypes(include="object")#if there is exict a column containing only one value
for i in df_num:
    if df_num[i].var()==0:
        del df_num[i]
cols_should_dropped=[i for i in df_object.columns if len(df_object[i].unique())==1]
df_object=df_object.drop(columns=cols_should_dropped)
```

```python
from scipy.stats.mstats import winsorize
```

```python
dataframe=data.copy()
```

```python
dataframe.info()
```

```python
dataframe["period"] = dataframe["brand"].map(str) + dataframe["model"]
```

```python
## brands=data.brand.unique()
models=dataframe.period.unique()
```

```python
for i in models:
    dataframe.loc[dataframe["period"]==i,"price"]=winsorize(dataframe.loc[dataframe["period"]==i,"price"],inplace=True,limits=0.10)
```

```python
plt.boxplot(dataframe.price)
```

```python
dataframe[(dataframe.price>47200) & (dataframe.Horsepower>450)]
```

```python
dataframe=dataframe.drop(dataframe[dataframe["price"]>50000].index)
```

```python
dataframe.reset_index(drop=True,inplace=True)
```

```python
dataframe=dataframe.drop_duplicates(subset="id_val")
dataframe.reset_index(drop=True,inplace=True)
```
